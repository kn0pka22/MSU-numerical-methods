```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filename = "p.txt" 

data = pd.read_csv(filename, sep='\s+', header=0, names=['N', 'Quadrature', 'Analytical', 'error'])
N          = data['N']
Quadrature = data['Quadrature'] 
Analytical = data['Analytical']
error      = data['error']

#data.head()

plt.figure(figsize=(10, 6))
plt.plot(np.log(N), np.log(error), label='error', marker='o', linestyle='-', color='b')


plt.xlabel('log(N)', fontsize=14)
plt.ylabel('log(err)', fontsize=14)
plt.xscale('log') 
plt.title('convergence', fontsize=16)
plt.legend()

plt.grid(True)

plt.savefig(f"convergence.png")  
plt.show()
```
