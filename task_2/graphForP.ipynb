```python
import numpy as np
import matplotlib.pyplot as plt

N = 10 
testfilename = 'PDependence.txt'  
plt.figure(figsize=(10, 10))  
plt.title(f"{testfilename} - {N} knots", fontsize=20, fontweight='bold') 
plt.grid(True) 


data = np.loadtxt(testfilename)  
x = data[:, 0]  #data[0] 
y = data[:, 1]  #data[1]

plt.plot(x, y, color='red', linewidth=1, marker='o', label='Dependence -log(norm) of -log(h)')


plt.xlabel('-log(h)', fontsize=16)
plt.ylabel('-log(norm)', fontsize=16)


plt.savefig(f"PDependence.png")  # Сохранение в формате PNG
plt.show()  

```
