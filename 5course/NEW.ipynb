

Решается параболическое уравнение в одномерном случае:

$$
u_t = u_{xx} + b \cdot u, \quad b > 0, \quad x \in (0, 1)
$$

С граничными условиями :
$$
u(t, 0) = u(t, 1) = 0
$$

И начальным условием:
$$
u(0, x) = \sin(\pi x)
$$

### 2. Численный метод (неявная разностная схема)
Используется **неявная разностная схема** первого порядка по времени и второго порядка по пространству:

$$
\frac{u_m^{n+1} - u_m^n}{\tau} = \frac{u_{m+1}^{n+1} - 2u_m^{n+1} + u_{m-1}^{n+1}}{h^2} + b \cdot u_m^{n+1}
$$

- $ \tau $ — шаг по времени  
- $ h $ — шаг по пространству  
- $ u_m^n $ — значение решения в узле $ x_m $ на временном слое $ n $

Эта схема абсолютно устойчива, что позволяет использовать большие шаги по времени.

### 3. Решение системы уравнений (метод прогонки)
На каждом временном слое получается **система линейных уравнений** с трёхдиагональной матрицей:

$$
A \cdot u^{n+1} = u^n
$$

где $ A $ — трёхдиагональная матрица.

Для её решения применяется **метод прогонки (алгоритм Томаса)**, который имеет сложность $ O(M) $ и не требует хранения всей матрицы.

### 4. Исследование устойчивости
Собственные значения оператора $ \frac{\partial^2}{\partial x^2} + b $:

$$
\mu_k = b - \pi^2 k^2
$$

- Если $ b > \pi^2 $, то $ \mu_1 > 0 $ — решение будет расти экспоненциально.
- Если $ b < \pi^2 $, то все $ \mu_k < 0 $ — решение будет затухать.

В задаче рассматривается два случая:
1. $ b = 12 $ (рост, т.к. $ 12 > \pi^2 \approx 9.87 $)
2. $ b = 2 $ (затухание, т.к. $ 2 < \pi^2 $)




```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 12})

def solveTridiagonalSystem(lowerDiag, mainDiag, upperDiag, rightSide):
    """
    Решает трёхдиагональную систему методом прогонки (алгоритм Томаса).
    
    Система уравнений имеет вид:
    a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1] = d[i]
    
    Параметры:
    lowerDiag - нижняя диагональ (коэффициенты a[i], длина n)
    mainDiag - главная диагональ (коэффициенты b[i], длина n)
    upperDiag - верхняя диагональ (коэффициенты c[i], длина n)
    rightSide - правая часть (коэффициенты d[i], длина n)
    
    Возвращает:
    solution - решение системы (длина n)
    """
    n = len(rightSide)
    
    print(f"Начинаю решение трёхдиагональной системы размером {n}...")
    
    alpha = np.zeros(n - 1)
    beta = np.zeros(n)
    
    # Начальные значения
    alpha[0] = -upperDiag[0] / mainDiag[0]
    beta[0] = rightSide[0] / mainDiag[0]
    
    for i in range(1, n - 1):
        denominator = mainDiag[i] + lowerDiag[i] * alpha[i - 1]
        alpha[i] = -upperDiag[i] / denominator
        beta[i] = (rightSide[i] - lowerDiag[i] * beta[i - 1]) / denominator
    
    # Последний элемент
    denominator = mainDiag[n - 1] + lowerDiag[n - 1] * alpha[n - 2]
    beta[n - 1] = (rightSide[n - 1] - lowerDiag[n - 1] * beta[n - 2]) / denominator
    
    # Обратный ход: нахождение решения
    solution = np.zeros(n)
    solution[n - 1] = beta[n - 1]
    
    for i in range(n - 2, -1, -1):
        solution[i] = alpha[i] * solution[i + 1] + beta[i]
    
    print(f"Решение трёхдиагональной системы завершено.")
    return solution


def createComputationalGrid(finalTime, timeSteps, length, spaceSteps):
    """
    Создаёт равномерную сетку для решения уравнения.
    
    Параметры:
    finalTime - конечное время моделирования T
    timeSteps - количество шагов по времени N
    length - длина области по пространству L
    spaceSteps - количество шагов по пространству M
    
    Возвращает:
    xNodes - узлы сетки по пространству
    tNodes - узлы сетки по времени
    dx - шаг по пространству h
    dt - шаг по времени τ
    """
    dx = length / spaceSteps
    dt = finalTime / timeSteps
    
    xNodes = np.linspace(0, length, spaceSteps + 1)
    tNodes = np.linspace(0, finalTime, timeSteps + 1)
    
    print("\n" + "="*60)
    print("ИНФОРМАЦИЯ О ВЫЧИСЛИТЕЛЬНОЙ СЕТКЕ:")
    print("="*60)
    print(f"Длина области: L = {length}")
    print(f"Конечное время: T = {finalTime}")
    print(f"Шаг по пространству: h = {dx:.6f}")
    print(f"Шаг по времени: τ = {dt:.6f}")
    print(f"Узлов по пространству: {len(xNodes)}")
    print(f"Узлов по времени: {len(tNodes)}")
    print(f"Число Куранта: τ/h² = {dt/(dx**2):.4f}")
    
    return xNodes, tNodes, dx, dt

# РЕШЕНИЕ ПАРАБОЛИЧЕСКОГО УРАВНЕНИЯ
def solveHeatEquation(initialConditionFunc, bParam, length, spaceSteps, finalTime, timeSteps):
    """
    Решает параболическое уравнение u_t = u_xx + b*u.
    
    Параметры:
    initialConditionFunc - функция начального условия
    bParam - коэффициент источника b
    length, spaceSteps - параметры пространственной области
    finalTime, timeSteps - параметры временного интервала
    
    Возвращает:
    solutionMatrix - матрица решения (размер [timeSteps+1, spaceSteps+1])
    xNodes, tNodes - координаты узлов сетки
    """
    
    print("\n" + "="*60)
    print(f"РЕШЕНИЕ ПАРАБОЛИЧЕСКОГО УРАВНЕНИЯ С b = {bParam}")
    print("="*60)
    
    # Создаём сетку
    xNodes, tNodes, dx, dt = createComputationalGrid(finalTime, timeSteps, length, spaceSteps)
    
    # Инициализируем матрицу решения
    solutionMatrix = np.zeros((timeSteps + 1, spaceSteps + 1))
    
    # Задаём начальное условие
    print("\nЗадаю начальное условие u(0, x) = sin(πx)...")
    solutionMatrix[0, :] = initialConditionFunc(xNodes)
    
    # Применяем граничные условия
    solutionMatrix[0, 0] = 0
    solutionMatrix[0, -1] = 0
    
    print(f"Начальные значения: min = {np.min(solutionMatrix[0, :]):.4f}, max = {np.max(solutionMatrix[0, :]):.4f}")
    
    # Подготовка коэффициентов для метода прогонки
    innerPoints = spaceSteps - 1  # Внутренние узлы (без границ)
    
    # Коэффициенты трёхдиагональной матрицы
    diagMain = 1/dt - bParam + 2/(dx**2)
    diagSide = -1/(dx**2)
    
    print(f"\nКОЭФФИЦИЕНТЫ РАЗНОСТНОЙ СХЕМЫ:")
    print(f"Главная диагональ: {diagMain:.4f}")
    print(f"Побочные диагонали: {diagSide:.4f}")
    
    # Формируем диагонали матрицы
    lowerDiagonal = np.full(innerPoints, diagSide)
    mainDiagonal = np.full(innerPoints, diagMain)
    upperDiagonal = np.full(innerPoints, diagSide)
    
    # Временной цикл
    print(f"\nНачинаю временные итерации...")
    for timeIndex in range(timeSteps):
        # Формируем правую часть системы
        rightHandSide = (1/dt) * solutionMatrix[timeIndex, 1:-1]
        
        # Решаем систему методом прогонки
        innerSolution = solveTridiagonalSystem(lowerDiagonal, mainDiagonal, upperDiagonal, rightHandSide)
        
        # Сохраняем решение для внутренних узлов
        solutionMatrix[timeIndex + 1, 1:-1] = innerSolution
        
        # Граничные условия на каждом временном слое
        solutionMatrix[timeIndex + 1, 0] = 0
        solutionMatrix[timeIndex + 1, -1] = 0
        
        # Вывод прогресса каждые 10% итераций
        if (timeIndex + 1) % max(1, timeSteps // 10) == 0:
            currentTime = tNodes[timeIndex + 1]
            currentMax = np.max(solutionMatrix[timeIndex + 1, :])
            print(f"  Шаг {timeIndex + 1}/{timeSteps}, t = {currentTime:.4f}, max(u) = {currentMax:.4f}")
    
    # Анализ результатов
    finalMax = np.max(solutionMatrix[-1, :])
    finalMin = np.min(solutionMatrix[-1, :])
    print(f"\nРЕЗУЛЬТАТЫ РАСЧЁТА:")
    print(f"Максимальное значение в конце: {finalMax:.6f}")
    print(f"Минимальное значение в конце: {finalMin:.6f}")
    print(f"Изменение максимума: {finalMax - np.max(solutionMatrix[0, :]):.6f}")
    
    return solutionMatrix, xNodes, tNodes


def visualizeSolution(solutionMatrix, xNodes, tNodes, bValue, caseDescription):
    """
    Визуализирует результаты расчёта.
    
    Параметры:
    solutionMatrix - матрица решения
    xNodes, tNodes - координаты узлов сетки
    bValue - значение параметра b
    caseDescription - описание случая для заголовков
    """
    
    print(f"\nСоздаю графики для случая b = {bValue}...")
    
    fig = plt.figure(figsize=(15, 10))
    
    ax1 = plt.subplot(2, 2, 1)
    
    timeIndices = [0, len(tNodes)//4, len(tNodes)//2, 3*len(tNodes)//4, -1]
    colors = ['blue', 'green', 'orange', 'red', 'purple']
    
    for idx, color in zip(timeIndices, colors):
        ax1.plot(xNodes, solutionMatrix[idx, :], 
                color=color, linewidth=2, 
                label=f't = {tNodes[idx]:.3f}')
    
    ax1.set_xlabel('Координата x', fontsize=12)
    ax1.set_ylabel('Решение u(x,t)', fontsize=12)
    ax1.set_title(f'Профили решения ({caseDescription})', fontsize=14, fontweight='bold')
    ax1.grid(True, linestyle='--', alpha=0.5)
    ax1.legend(loc='best')
    
    # 2. L2-норма решения как функция времени
    ax2 = plt.subplot(2, 2, 2)
    
    # Вычисляем L2-норму (энергию) на каждом временном слое
    dx = xNodes[1] - xNodes[0]
    l2Norm = np.sqrt(np.sum(solutionMatrix**2, axis=1) * dx)
    
    ax2.plot(tNodes, l2Norm, 'b-', linewidth=2)
    ax2.set_xlabel('Время t', fontsize=12)
    ax2.set_ylabel('L2-норма ||u(t)||', fontsize=12)
    ax2.set_title(f'Эволюция L2-нормы ({caseDescription})', fontsize=14, fontweight='bold')
    ax2.grid(True, linestyle='--', alpha=0.5)
    
    # 3. 3D-поверхность решения
    ax3 = plt.subplot(2, 2, (3, 4), projection='3d')
    
    # Создаём сетку для 3D графика
    X, T = np.meshgrid(xNodes, tNodes)
    
    # Рисуем поверхность
    surf = ax3.plot_surface(X, T, solutionMatrix, cmap='viridis', 
                           alpha=0.9, linewidth=0.1, antialiased=True)
    
    ax3.set_xlabel('Координата x', fontsize=10)
    ax3.set_ylabel('Время t', fontsize=10)
    ax3.set_zlabel('Решение u(x,t)', fontsize=10)
    ax3.set_title(f'Решение u(x,t) в 3D ({caseDescription})', fontsize=14, fontweight='bold')
    
    # Добавляем цветовую шкалу
    fig.colorbar(surf, ax=ax3, shrink=0.6, aspect=10, label='Значение u')
    
    plt.suptitle(f'Параболическое уравнение: u_t = u_xx + {bValue}·u', 
                fontsize=16, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.show()
    
    # Дополнительный анализ
    print("\nАНАЛИЗ РЕЗУЛЬТАТОВ:")
    print(f"Начальная L2-норма: {l2Norm[0]:.6f}")
    print(f"Конечная L2-норма: {l2Norm[-1]:.6f}")
    print(f"Изменение L2-нормы: {l2Norm[-1] - l2Norm[0]:.6f}")
    
    if l2Norm[-1] > l2Norm[0]:
        print("ЗАКЛЮЧЕНИЕ: Решение растёт во времени (b > π²)")
    else:
        print("ЗАКЛЮЧЕНИЕ: Решение затухает во времени (b < π²)")

# ============================================================================
# 5. ОСНОВНАЯ ПРОГРАММА
# ============================================================================
def main():
    """
    Основная функция программы.
    Решает параболическое уравнение для двух случаев.
    """
    
    print("\n" + "="*60)
    print("ЗАДАЧА УПРАВЛЕНИЯ ДЛЯ ПАРАБОЛИЧЕСКОГО УРАВНЕНИЯ")
    print("Уравнение: u_t = u_xx + b·u, x ∈ (0, 1), b = const > 0")
    print("Граничные условия: u(t, 0) = u(t, 1) = 0")
    print("Начальное условие: u(0, x) = sin(πx)")
    print("="*60)
    
    # Параметры расчёта
    L = 1.0          # Длина области
    M = 50           # Число узлов по пространству
    T = 0.5          # Конечное время
    N = 200          # Число шагов по времени
    
    print(f"\nОБЩИЕ ПАРАМЕТРЫ РАСЧЁТА:")
    print(f"Длина области: L = {L}")
    print(f"Узлов по пространству: M = {M}")
    print(f"Конечное время: T = {T}")
    print(f"Шагов по времени: N = {N}")
    
    # Определяем начальное условие
    def initialCondition(x):
        return np.sin(np.pi * x)
    
    # СЛУЧАЙ 1: b = 12 (рост решения, т.к. 12 > π² ≈ 9.87)
    print("\n" + "="*60)
    print("СЛУЧАЙ 1: ИССЛЕДОВАНИЕ РОСТА РЕШЕНИЯ")
    print("Параметр b = 12 > π², ожидается рост решения")
    print("="*60)
    
    b1 = 12.0
    solution1, xNodes1, tNodes1 = solveHeatEquation(
        initialCondition, b1, L, M, T, N
    )
    
    visualizeSolution(solution1, xNodes1, tNodes1, b1, "рост при b=12")
    
    # СЛУЧАЙ 2: b = 2 (затухание решения, т.к. 2 < π²)
    print("\n" + "="*60)
    print("СЛУЧАЙ 2: ИССЛЕДОВАНИЕ ЗАТУХАНИЯ РЕШЕНИЯ")
    print("Параметр b = 2 < π², ожидается затухание решения")
    print("="*60)
    
    b2 = 2.0
    solution2, xNodes2, tNodes2 = solveHeatEquation(
        initialCondition, b2, L, M, T, N
    )
    
    visualizeSolution(solution2, xNodes2, tNodes2, b2, "затухание при b=2")
    
    # Сравнение двух случаев
    print("\n" + "="*60)
    print("СРАВНИТЕЛЬНЫЙ АНАЛИЗ ДВУХ СЛУЧАЕВ")
    print("="*60)
    
    # Вычисляем L2-нормы для сравнения
    dx = xNodes1[1] - xNodes1[0]
    l2Norm1 = np.sqrt(np.sum(solution1**2, axis=1) * dx)
    l2Norm2 = np.sqrt(np.sum(solution2**2, axis=1) * dx)
    
    # Создаём сравнительный график
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # График L2-норм
    axes[0].plot(tNodes1, l2Norm1, 'r-', linewidth=2, label='b = 12 (рост)')
    axes[0].plot(tNodes2, l2Norm2, 'b-', linewidth=2, label='b = 2 (затухание)')
    axes[0].set_xlabel('Время t', fontsize=12)
    axes[0].set_ylabel('L2-норма ||u(t)||', fontsize=12)
    axes[0].set_title('Сравнение L2-норм для разных b', fontsize=14, fontweight='bold')
    axes[0].grid(True, linestyle='--', alpha=0.5)
    axes[0].legend()
    
    # График конечных профилей
    axes[1].plot(xNodes1, solution1[-1, :], 'r-', linewidth=2, label='b = 12 (конечный профиль)')
    axes[1].plot(xNodes2, solution2[-1, :], 'b-', linewidth=2, label='b = 2 (конечный профиль)')
    axes[1].plot(xNodes1, solution1[0, :], 'r--', linewidth=1, alpha=0.5, label='Начальное условие')
    axes[1].set_xlabel('Координата x', fontsize=12)
    axes[1].set_ylabel('Решение u(x,T)', fontsize=12)
    axes[1].set_title('Сравнение конечных профилей', fontsize=14, fontweight='bold')
    axes[1].grid(True, linestyle='--', alpha=0.5)
    axes[1].legend()
    
    plt.suptitle('Сравнение поведения решения для разных значений b', 
                fontsize=16, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.show()
    
    # Вывод итоговых результатов
    print("\n" + "="*60)
    print("ИТОГОВЫЕ РЕЗУЛЬТАТЫ:")
    print("="*60)
    print(f"Случай b = 12.0:")
    print(f"  Начальная L2-норма: {l2Norm1[0]:.6f}")
    print(f"  Конечная L2-норма:  {l2Norm1[-1]:.6f}")
    print(f"  Отношение конечной к начальной: {l2Norm1[-1]/l2Norm1[0]:.6f}")
    print(f"  Максимальное значение в конце: {np.max(solution1[-1, :]):.6f}")
    
    print(f"\nСлучай b = 2.0:")
    print(f"  Начальная L2-норма: {l2Norm2[0]:.6f}")
    print(f"  Конечная L2-норма:  {l2Norm2[-1]:.6f}")
    print(f"  Отношение конечной к начальной: {l2Norm2[-1]/l2Norm2[0]:.6f}")
    print(f"  Максимальное значение в конце: {np.max(solution2[-1, :]):.6f}")
    
    print("\n" + "="*60)
    print("ПРОГРАММА ЗАВЕРШЕНА УСПЕШНО!")
    print("="*60)

# Запуск основной программы
if __name__ == "__main__":
    main()
```
