```python
a = 1
a
import numpy as np
```

```python
def func(x1, x2):
    b = -(x1+x2)
    c = x1*x2

    D = b*b - 4*a*c
    # print(D)
    root1 = (-b+np.sqrt(D))/2.
    root2 = (-b-np.sqrt(D))/2.
    return root1, root2

def f(x, x1, x2):
    return x*x - (x1+x2)*x+x1*x2



print("         n               x1                       x2                      f(x1)                      f(x2)")

for n in range(0, 14):
    root1, root2 = func(10**n, 10**(-n))
    print(f"{n:10.2F}", '    ',f"{root1:.14E}", '    ',f"{root2:.14E}", '    ',f"{f(root1, 10**n,10**(-n)):.14E}", '    ',f"{f(root2,10**n,10**(-n)):.14E}")


```
