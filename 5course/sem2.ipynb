```python
import numpy as np
import matplotlib.pyplot as plt

def int_sum_2n():
    S = 0
    Sn = 0
    qn = 1
    while True:
        Sn = S + qn
        if S == Sn:
            break
        qn = qn * 2
        S = Sn
    return S

def deg(x, n):
    res = 1
    for i in range(n):
        res *= x
    return res

def func(x):
    # (x-2)^9 = x^9 - 18x^8 + 144x^7 - 672x^6 + 2016x^5 - 4032x^4 + 5376x^3 - 4608x^2 + 2304x - 512
    return (deg(x, 9) - 18*deg(x, 8) + 144*deg(x, 7) - 672*deg(x, 6) + 
            2016*deg(x, 5) - 4032*deg(x, 4) + 5376*deg(x, 3) - 4608*deg(x, 2) + 
            2304*x - 512)

result = int_sum_2n()
print(f"Результат IntSum2n(): {result}")

x = np.linspace(1, 3, 100)  
y = func(x)  

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('График функции (x-2)^9')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=2, color='r', linestyle='--', alpha=0.7, label='x=2')
plt.legend()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
expr = (x-2)**9
expanded_expr = sp.expand(expr)

print("Исходное выражение:", expr)
print("Разложенное выражение:", expanded_expr)

f_expanded = sp.lambdify(x, expanded_expr, 'numpy')

x_vals = np.linspace(1.999, 2.001, 1000)
y_vals = f_expanded(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'darkred', linewidth=2, alpha=0.8)

plt.style.use('seaborn-v0_8-whitegrid')
plt.title('График разложенной формы $(x-2)^9$', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11, frameon=True, fancybox=True, shadow=True)

plt.axvline(x=2, color='gray', linestyle='--', alpha=0.7, label='x=2')

plt.tight_layout()
plt.show()
```
