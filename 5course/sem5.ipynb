```python
import math

def coef(alpha, k):
    coef = 1.0
    for i in range(1, k + 1):
        coef *= (alpha - i + 1) / i
    return coef

def taylor_root(a, m, n_terms, max_iter=100, tol=1e-14):
    """
    x^m - a = 0 
    Parameters:
    a - число, из которого извлекаем корень
    m - степень корня
    n_terms - количество слагаемых в разложении Тейлора
    max_iter - максимальное количество итераций
    tol - точность
    
    Returns:
    корень и количество итераций
    """

    # Начальное приближение
    xn = 1.0 if a > 0 else -1.0
    
    for iter_count in range(1, max_iter + 1):
        # Вычисляем f(xn) = xn^m - a
        f_xn = xn**m - a
        
        if abs(f_xn) < tol:
            return xn, iter_count
        
        delta = 0.0
        
        for k in range(1, n_terms + 1):
            # k-я производная обратной функции в точке y0 = f(xn)
            # F^{(k)}(y0) = (1/m)(1/m-1)...(1/m-k+1) * (y0+a)^{1/m - k}
            # При y0 = xn^m - a, (y0+a) = xn^m
            term = coef(1/m, k) * (-f_xn)**k / xn**(m*k - 1)
            delta += term
        
        xn_new = xn + delta
        
        if abs(xn_new - xn) < tol:
            return xn_new, iter_count
        
        xn = xn_new
    
    return xn, max_iter

test_cases = [
    (2, 2),   # квадратный корень из 2
    (3, 8),   # кубический корень из 8  
    (5, 32),  # корень 5-й степени из 32
    (2, 9),   # квадратный корень из 9
    (4, 16),  # корень 4-й степени из 16
]

print("Сравнение скорости сходимости для разных количеств слагаемых:")
print("=" * 70)

for m, a in test_cases:
    exact = a**(1/m)
    print(f"\nРешение x^{m} - {a} = 0 (точный корень: {exact:.10f})")
    print("-" * 50)
    
    for n in range(1, 6):  # от 1 до 5 слагаемых
        root, iterations = taylor_root(a, m, n)
        error = abs(root - exact)
        print(f"n={n}: {root:.10f}, итераций: {iterations}, ошибка: {error:.2e}")
```
